```python
def img_writer(img, boundary, dict,i):
    img_write_path = './exported_imgs/img_'+str(i)+'.jpg'
    xml_write_path = './xml/img_'+str(i)+'.xml'
    cv2.imwrite(img_write_path,img)
    dict['annotation']['filename'] = 'img_'+str(i)+'.jpg'

    z=0
    if len(boundary.bounding_boxes) >1:
        for item in boundary.bounding_boxes:
            print(item)
            dict['annotation']['object'][z]['bndbox']['xmin']=int(item.x1)
            dict['annotation']['object'][z]['bndbox']['xmax']=int(item.x2)
            dict['annotation']['object'][z]['bndbox']['ymin']=int(item.y1)
            dict['annotation']['object'][z]['bndbox']['ymax']=int(item.y2)
            z+=1
    else:    
        [data] = boundary.bounding_boxes
        dict['annotation']['object']['bndbox']['xmin']=int(data.x1)
        dict['annotation']['object']['bndbox']['xmax']=int(data.x2)
        dict['annotation']['object']['bndbox']['ymin']=int(data.y1)
        dict['annotation']['object']['bndbox']['ymax']=int(data.y2)

    xml = dicttoxml.dicttoxml(dict,attr_type=False)
    file = open(xml_write_path, 'wb')
    file.write(xml)
    file.close()
```
